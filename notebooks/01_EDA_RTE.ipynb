# Analyse Exploratoire des Données RTE (PowerCast)
Ce notebook permet d'inspecter les données brutes téléchargées, de traiter les valeurs manquantes et d'observer les premières tendances temporelles (saisonnalités, pics hivernaux).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
data_path = '../data/raw/eco2mix_national.csv'
if os.path.exists(data_path):
    df = pd.read_csv(data_path, sep=';', low_memory=False)
    print(f"Dimensions du dataset : {df.shape}")
else:
    print("Veuillez exécuter src/data/download_rte_data.py en premier.")
```

```python
if 'df' in locals():
    display(df.head())
    print(df.columns.tolist())
```

```python
if 'df' in locals():
    # Traitement des colonnes temporelles
    df['Date et Heure'] = pd.to_datetime(df['Date et Heure'], utc=True)
    df = df.sort_values('Date et Heure').reset_index(drop=True)
    
    # Tracé de la consommation globale
    plt.figure(figsize=(15, 6))
    plt.plot(df['Date et Heure'], df['Consommation (MW)'], lw=0.5)
    plt.title('Consommation électrique nationale (RTE)')
    plt.xlabel('Date')
    plt.ylabel('Consommation (MW)')
    plt.show()
```
